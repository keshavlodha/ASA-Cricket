```python
import pandas as pd
from tqdm import tqdm
import pickle
import numpy as np
import os
```

```python
matches = pickle.load(open("dataset_lvl1.pkl", 'rb'))
```

```python
type(pd.DataFrame(matches))
```

```python
batting_team = (matches.iloc[65]['innings'][0])
batting_team
```

```python
delivery_df = pd.DataFrame()
for i in tqdm(range(len(matches))):

    if len(matches.iloc[i]['innings']) >= 2:
        deliveries = matches.iloc[i]['innings'][0]['1st innings']['deliveries']


        
        for delivery in (deliveries):
            #initialize temporary dictionary and list
            tmp_list = []
            tmp = {}

            #extract required data form the innings dictionary
            match_id = matches.iloc[i]['match_id']
            batting_team = matches.iloc[i]['innings'][0]['1st innings']['team']
            ball = list(delivery.keys())[0]
            batsman = delivery[ball]['batsman']
            bowler = delivery[ball]['bowler']
            city = matches.iloc[i]['info.city']
            venue = matches.iloc[i]['info.venue']
            batsman_runs = delivery[ball]['runs']['batsman']
            extras_runs = delivery[ball]['runs']['extras']

            if(delivery[ball].get('wicket') != None):
                player_dismissed = 1
            else:
                player_dismissed = 0

            #add extracted data into a temporary dictionary
            tmp['match_id'] = match_id
            tmp['batting_team'] = batting_team
            tmp['ball'] = ball
            tmp['batsman'] = batsman
            tmp['bowler'] = bowler
            tmp['batsman_runs'] = batsman_runs
            tmp['extras_runs'] = extras_runs
            tmp['player_dismissed'] = player_dismissed
            tmp['city'] = city
            tmp['venue'] = venue

            #convert the temp dictionary into a pandas df
            tmp_list.append(tmp)
            tmp_df = pd.DataFrame(tmp_list)

            #add temp df to delivery df
            delivery_df = pd.concat([delivery_df, tmp_df], ignore_index=True)

```
