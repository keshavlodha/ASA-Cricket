```python
import pandas as pd
import numpy as np
import os
from tqdm import tqdm
from yaml import safe_load
import pickle
```

```python
filenames = []
for file in os.listdir('t20s'):
    filenames.append(os.path.join('t20s', file))
```

```python
print(filenames)
```

```python
final_df = pd.DataFrame()
cnt = 1
error_terms = 0
for file in tqdm(filenames):
    with open(file, 'r') as f:
        try:
            df = pd.json_normalize(safe_load(f))
        except:
            error_terms += 1
            continue
        df['match_id'] = cnt
        final_df = pd.concat([final_df, df], ignore_index=True)
        cnt+=1
```

```python
backup_df = final_df.copy()
```

```python
backup_df.to_csv("initialData.csv")
```

```python
# Filtering to account for Men's matches only as there is more data
final_df = final_df[final_df['info.gender'] == 'male']
```

```python
# Deleting unnecessary columns
final_df.drop(columns=['meta.data_version', 
                       'meta.created',
                       'meta.revision',
                       'info.gender',
                       'info.player_of_match',
                       'info.umpires',
                       'info.bowl_out',
                       'info.outcome.bowl_out',
                       'info.supersubs.New Zealand',
                       'info.supersubs.South Africa',
                       'info.outcome.winner',
                       'info.outcome.by.wickets',
                       'info.neutral_venue',
                       'info.outcome.by.runs',
                      'info.outcome.method',
                      'info.outcome.eliminator',
                      'info.outcome.result',
                      'info.match_type_number',
                      'info.outcome.by.wickets'], inplace=True)
```
